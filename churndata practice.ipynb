```python
#Transforming churn data to 0s and 1s
Yesdf = data[data['churn'] =='no']
Yesdf['churn'] = 0
Nodf = data[data['churn'] == 'yes']
Nodf['churn'] = 1
newdata = pd.concat([Yesdf, Nodf])
```

```python
#Transforming international plan to 0s and 1s
data1 = newdata[newdata['international_plan'] == 'no']
data1['international_plan'] = 0
data2 = newdata[newdata['international_plan'] == 'yes']
data2['international_plan'] = 1
data3 = pd.concat([data1, data2])
```
